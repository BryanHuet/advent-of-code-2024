### Part 1


```python
def walk_until_obstacle(position, maps, obstacle = '#', mark = 'X'):
    coord = position[0]
    direction = position[1]
    limit = (len(maps), len(maps[0]))

    if direction=='^' and coord[0] <= 0:
        return maps
    if direction=='<' and coord[1] <=0:
        return maps
    if direction=='>' and coord[1] >= limit[1]:
        return maps

    if direction == '^':
        if coord[0] <= 0:
            maps[0][coord[1]]=mark
            return maps
        for i in range(coord[0]-1, -1,-1):
            if maps[i][coord[1]] == obstacle:
                return maps
            maps[i+1][coord[1]] = mark
            maps[i][coord[1]] = '>'
            if i == 0:
                maps[i][coord][1]=mark
                return maps        
        return maps

    if direction == '>':
            if coord[1] >= limit[1]:
                maps[coord[0]][coord[1]]=mark
                return maps
            for j in range(coord[1]+1, limit[1]):
                if maps[coord[0]][j] == obstacle:
                    return maps
                maps[coord[0]][j-1] = mark
                maps[coord[0]][j] = 'v'
                if j == limit[1]-1:
                    maps[coord[0]][j]=mark
                    return maps             
            return maps

    if direction == 'v':
            if coord[0] >= limit[1]:
                maps[coord[0]][coord[1]]=mark
                return maps
            for i in range(coord[0]+1, limit[0]):
                if maps[i][coord[1]] == obstacle:
                    return maps
                maps[i-1][coord[1]] = mark
                maps[i][coord[1]] = '<'  
                if i == limit[0]-1:
                    maps[i][coord[1]]=mark
                    return maps        
            return maps

    if direction == '<':
            if coord[1] <= 0:
                maps[coord[0]][coord[1]]=mark
                return maps
            for j in range(coord[1]-1, -1,-1):
                if maps[coord[0]][j] == obstacle:
                    return maps
                maps[coord[0]][j+1] = mark
                maps[coord[0]][j] = '^'    
                if j == 0:
                    maps[coord[0]][j]=mark
                    return maps      
            return maps

def find_position(maps):
    for i in range(len(maps)):
        for j in range(len(maps[0])):
            if maps[i][j]=='^':
                return (i,j), '^' 
            if maps[i][j]=='>':
                return (i,j), '>' 
            if maps[i][j]=='<':
                return (i,j), '<' 
            if maps[i][j]=='v':
                return (i,j), 'v' 

def is_there(maps):
    return 'v' in maps or '>' in maps or '<' in maps or '^' in maps

def process_patrol(filename, print_map=False):
    with open(filename, 'r') as file:
        maps = [ list(line) for line in file.read().split('\n')]

    position = find_position(maps)
    if print_map:
        print('\n'.join([''.join(line) for line in maps]))
        print('\n')
    
    maps_str = '\n'.join([''.join(line) for line in maps])
    while(is_there(maps_str)):
        position = find_position(maps)
        maps = walk_until_obstacle(position, maps)
        maps_str = '\n'.join([''.join(line) for line in maps])

    return maps_str.count('X')

# print('\n'.join([''.join(line) for line in process_patrol('test.txt', True)]))
print('The guard will visit',process_patrol('patrol.txt'), 'distinct positions before leaving the area.')
```
